```python
import matplotlib.pyplot as plt
# Data
subjects = ["Math", "Physics", "Chemistry", "Biology",
"English"]
marks = [85, 90, 78, 88, 76]
# Plot
plt.plot(subjects, marks, marker='o', color='black')
# Customization
plt.title("Student Marks in Different Subjects")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.grid(True)
# Show plot
plt.show()

```

```python
import matplotlib.pyplot as plt
# Data
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
temperature = [30, 32, 31, 29, 33, 35, 34]
# Plot
plt.plot(days, temperature, marker='s', linestyle='--',
color='black')
# Customization
plt.title("Weekly Temperature Trend")
plt.xlabel("Days")
plt.ylabel("Temperature (Â°C)")
plt.grid(True)
# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
# Data
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales = [15000, 18000, 12000, 22000, 24000, 20000]
# Plot
plt.bar(months, sales, color='black')
# Customization
plt.title("Monthly Sales Report")
plt.xlabel("Months")
plt.ylabel("Sales in USD")
plt.grid(axis='y')
# Show plot
plt.show()
```
